# Get

```python
import requests
```

```python
res = requests.get("http://search.books.com.tw/search/query/key/swift/cat/all")
print(res.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.text, "html.parser")
print(soup.title.string)
```

```python
soup.select("img[class='itemcov']")
```

```python
print(type(soup.select("img[class='itemcov']")))

print(type(soup.select("img[class='itemcov']")[0]))
print(soup.select("img[class='itemcov']")[0])

print(type(soup.select("img[class='itemcov']")[0]['alt']))
print(soup.select("img[class='itemcov']")[0]['alt'])
```

```python
soup.select("span[class='price']")
```

```python
import pandas as pd
```

# Get book name

```python
#get book name
books = pd.Series()
for book in soup.select("img[class='itemcov']"):
    books = books.append(pd.Series([book['alt']])).reset_index(drop=True)
books
```

# Get price

```python
for price in soup.select("span[class='price']"):
    print(price.select('b')) # 打折數 價格
```

```python
# do get price
i = 0
prices = pd.Series()
for price in soup.select("span[class='price']"):
    if(i<books.size):
        prices = prices.append(pd.Series([price.select('b')[-1].string])).reset_index(drop=True)
    i+=1
print(prices)
```

# Combine to a DataFrame

```python
df = pd.DataFrame({'Name':books, 'Price':prices})
df
```

# Post

```python
payload = {
    "FromCity":"0",
    "FromStation":"1008",
    "FromStationName":"0",
    "ToCity":"10",
    "ToStation":"1238",
    "ToStationName":"0",
    "TrainClass":"2",
    "searchdate":"2017-10-04",
    "FromTimeSelect":"0000",
    "ToTimeSelect":"2359",
    "Timetype":"1"
}
res_post = requests.post("http://twtraffic.tra.gov.tw/twrail/TW_SearchResult.aspx",data = payload)
print(res_post.text)
```

```python
soup_post = BeautifulSoup(res_post.text, 'html.parser')
print(soup_post.title.string)
```

```python
for train in soup_post.select("script"):
    print(train)
```

```python
for train in soup_post.select("script"):
    if('var JSONData' in train.text):
        print(train.text)
        trains = train.text
```

```python
trains[13:-1]
```

```python
import json
result = json.loads(trains[13:-1])
result[0:10]
```

```python
df_post = pd.DataFrame(result)
df_post
```

# XML

```python
import xml.etree.ElementTree as et
res = requests.get("https://cloud.culture.tw/frontsite/trans/emapOpenDataAction.do?method=exportEmapXMLByMainType&mainType=10")
print(res.text)
```

```python
tree = et.ElementTree(et.fromstring(res.text))
root = tree.getroot()
root.tag, root.attrib
```

## All tag in tree

```python

for elem in tree.iter():
    print(elem.tag, elem.attrib)
```

```python
info = []
for elem in tree.iter(tag='Info'):
    print(elem.tag, elem.attrib)
    info.append(elem.attrib)
```

```python
print(type(info))
df_xml_info = pd.DataFrame(info)
```
